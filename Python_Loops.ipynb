```python
# for loops
for number in range(1, 10, 2):
  print("Attempt", number, number * ".")
```

```python
# for...else
successful = True
for number in range(3):
  print("Attempt")
  if successful:
    print("Successful")
    break
else:
  print("Attempted 3 times and failed")
```

```python
# nested loops
for x in range(5):
  for y in range(3):
    print(f"({x},{y})")
```

```python
# iterables

print(type(5))
print(type(range(5)))

for x in [1, 2, 3, 4]:
  print(x)
```

```python
# while loops
number = 100
while number > 0:
  print(number)
  number //= 2

command = ""
while command.lower() != "quit":
  command = input(">")
  print("ECHO", command)
```

```python
# infinite loops
command = ""
while True:
  command = input(">")
  print("ECHO", command)
  if command.lower() == "quit":
    break
```

```python
count = 0
for number in range(1,10):
  if number % 2 == 0:
    count += 1
    print(number)

print(f"We have {count} even numbers")
```
